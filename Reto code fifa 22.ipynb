Introduccion, en este codigo utilizaremos una base de datos de el juego de Fifa 22. En esta base de datos se tienen todos los jugadores de los 10 paises mas grandes en el futbol. En esta base de datos tenemos variables como overall, nationality, entre algunas otra variables las cuales vamosm a comparar para presentar los datos. Una de las cosas la cual se va a tratar de observar es cual es el pais con los jugadores mas altos overall, ademas de esto se utilizara la base de datos para presentar graficos entre algunos otros analisis para observar los datos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
import seaborn as sns
```

```python
df = pd.read_csv('FIFA 22 TOP 10 PAISES.csv')
df
```

```python
df.info()
```

En esta parte podemos observar los tipos de datos que tenemos ademas de las variables que existen en el documento, utilizando info() podemos encontrar esto. Viendo esto podemos ver que datos podemos utilizar y caules no.

```python
df.describe()
```

```python
df['overall'].plot(kind='hist',
                   figsize=(20,10))
```

```python
sns.scatterplot(x='age',y='potential',data=df)
plt.title('Edad y su Potencial')
plt.show()
```

```python
pd.pivot_table(df,index='nationality',
              values='overall',
              aggfunc = ['sum'])
```

```python
sns.boxplot(data=df.loc[:,['overall']])
```

```python
df.corr()
```

```python
sns.scatterplot(x='pace',y='overall',data=df)
plt.title('Velocidad y Calidas en general')
plt.show()
```

```python
sns.lmplot(x='overall',
          y='age',
          data=df)
```

```python
plt.figure(figsize=(5,5))
sns.catplot(x='nationality',
           y='overall',
           data=df,
           kind='bar')
plt.show
```

```python
nationality = df['nationality']
overall = df['overall']
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x = nationality, y = overall)
plt.xlabel('nationalities')
plt.ylabel('overall') 
plt.title('overalls per nationality')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='overall',
               y='nationality',
               hue='age',
               data=df)
plt.show
```
